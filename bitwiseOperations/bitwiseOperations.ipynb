Bitwise Operations on Images

```python
#Importing necessary libraries
import numpy as np
import pandas as pd
import cv2 as cv
from matplotlib import pyplot as plt
#Reading respective images
bl=cv.imread('/Users/chinmayeed/Documents/TataIQ_Intern/bitwiseOperations/black.jpg')
wh=cv.imread('/Users/chinmayeed/Documents/TataIQ_Intern/bitwiseOperations/white.jpg')
fig = plt.figure(figsize=(8,8))
fig.add_subplot(1,2,1)
plt.imshow(bl)
plt.title('Black')
fig.add_subplot(1,2,2)
plt.imshow(wh)
plt.title('White')
```

AND Operation

```python
a = cv.bitwise_and(bl, wh, mask=None)
plt.imshow(a)
plt.title('AND')
```

OR Operation

```python
o = cv.bitwise_or(bl, wh, mask=None)
plt.imshow(o)
plt.title('OR')
```

XOR Operation

```python
x = cv.bitwise_xor(bl, wh, mask=None)
plt.imshow(x)
plt.title('XOR')
```

NOT Operation

```python
nb = cv.bitwise_not(bl,mask=None)
nw = cv.bitwise_not(wh,mask=None)
fig = plt.figure(figsize=(8,8))
fig.add_subplot(1,2,1)
plt.imshow(nb)
plt.title('NOT Black')
fig.add_subplot(1,2,2)
plt.imshow(nw)
plt.title('NOT White')
```
