Math Operations

```python
#Import necessary libraries
import cv2 as cv
from matplotlib import pyplot as plt
#Read images for addition
m1=cv.imread('/Users/chinmayeed/Documents/TataIQ_Intern/mathOp/m1.jpg')
m2=cv.imread('/Users/chinmayeed/Documents/TataIQ_Intern/mathOp/m2.jpg')
fig= plt.figure(figsize=(8,8))
fig.add_subplot(1,2,1)
plt.imshow(m1)
plt.title('Image 1')
fig.add_subplot(1,2,2)
plt.imshow(m2)
plt.title('Image 2')

#Read images for substraction
m3=cv.imread('/Users/chinmayeed/Documents/TataIQ_Intern/bitwiseOperations/black.jpg')
m4=cv.imread('/Users/chinmayeed/Documents/TataIQ_Intern/bitwiseOperations/white.jpg')
fig = plt.figure(figsize=(8,8))
fig.add_subplot(1,2,1)
plt.imshow(m3)
plt.title('Image 3')
fig.add_subplot(1,2,2)
plt.imshow(m4)
plt.title('Image 4')
```

Weighted addition

```python
#Performing Addition
wa = cv.addWeighted(m1, 0.5, m2, 0.5, 0)
#Plotting figures
fig= plt.figure(figsize=(8,8))
fig.add_subplot(1,1,1)
plt.imshow(wa)
plt.title('Weighted Addition')
```

Subtraction

```python
#Performing subtraction
s = cv.subtract(m3,m4)
#Plotting figures
fig= plt.figure(figsize=(8,8))
fig.add_subplot(1,1,1)
plt.imshow(s)
plt.title('Subtraction')
```
