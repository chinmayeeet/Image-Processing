Image Filters using OpenCV

```python
#Importing all necessary libraries
import cv2 as cv
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

Median Blur\n
Removes salt and pepper noise from images

```python
#Load Image
noiseimg = cv.imread("/Users/chinmayeed/Documents/TataIQ_Intern/download.jpeg")
#Remove noise
nonoiseimg = cv.medianBlur(noiseimg, 7)
#Plot images
fig = plt.figure(figsize=(10, 10))
fig.add_subplot(1,2,1)
plt.imshow(noiseimg)
plt.title('Image with noise')
fig.add_subplot(1,2,2)
plt.imshow(nonoiseimg)
plt.title('Image without noise')

```

Gaussian Filter\n
Gaussian filter blurs the desired area and cuts the noise with higher frequencies

```python
#Read image
gimg = cv.imread("/Users/chinmayeed/Documents/TataIQ_Intern/gaus.jpeg")
#Convert to grayscale
gray = cv.cvtColor(gimg, cv.COLOR_BGR2GRAY)
#Apply Gaussian Filter
gblur = cv.GaussianBlur(gimg, (11,11), 0)
#Plot images
fig = plt.figure(figsize=(10, 10))
fig.add_subplot(1,3,1)
plt.imshow(gimg)
plt.title('Image without filter')
fig.add_subplot(1,3,2)
plt.imshow(gray)
plt.title('Image in grayscale')
fig.add_subplot(1,3,3)
plt.imshow(gblur)
plt.title('Image without noise')
```
