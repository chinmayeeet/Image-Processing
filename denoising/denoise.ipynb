DENOISING IMAGES

Here we use median blur to denoise this particular image

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
  
# Reading image from folder where it is stored
img = cv2.imread('train.jpeg')
  
# denoising of image saving it into dst image
dst = cv2.medianBlur(img, 9)
  
# Plotting of source and destination image
plt.subplot(121), plt.imshow(img), plt.title('Noisy Image')
plt.subplot(122), plt.imshow(dst), plt.title('Denoised Image')
  
plt.show()
```
