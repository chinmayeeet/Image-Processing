Edge Detection Techinques

```python
#Importing all the necessary libraries
import cv2 as cv
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Read and display the image
img = cv.imread("giza.jpg")
plt.imshow(img)
plt.title('Original Image')
```

SOBEL EDGE DETECTION

In this method, edges are detected when there is a jump in the value of the first derivative

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
blur = cv.GaussianBlur(gray, (3, 3), 0)
sobelx = cv.Sobel(src=blur, ddepth=cv.CV_64F, dx=1, dy=0, ksize=7) # Sobel Edge Detection on the X axis (used to detect vertical edges)
sobely = cv.Sobel(src=blur, ddepth=cv.CV_64F, dx=0, dy=1, ksize=7) # Sobel Edge Detection on the Y axis (used to detect horizontal edges)
sobelxy = cv.Sobel(src=blur, ddepth=cv.CV_64F, dx=1, dy=1, ksize=7) # Combined X and Y Sobel Edge Detection
fig = plt.figure(figsize=(10, 10))
fig.add_subplot(1,4,1)
plt.imshow(img)
plt.title('Original Image')
fig.add_subplot(1,4,2)
plt.imshow(sobelx)
plt.title('Sobel X')
fig.add_subplot(1,4,3)
plt.imshow(sobely)
plt.title('Sobel Y')
fig.add_subplot(1,4,4)
plt.imshow(sobelxy)
plt.title('Sobel XY')

```

LAPLACIAN EDGE DETECTION

In this method, edges are detected when the value of the second derivative crosses 0

```python
ddepth = cv.CV_16S
dst = cv.Laplacian(blur, ddepth, ksize=21)
fig = plt.figure(figsize=(10, 10))
fig.add_subplot(1,2,1)
plt.imshow(img)
plt.title('Original Image')
fig.add_subplot(1,2,2)
plt.imshow(dst)
plt.title('Laplacian')

```

CANNY EDGE DETECTION

In this medthod the following rules are taken into consideration while detecting edges:
if a pixel is higher than the upper threshold, then it is accepted
if a pixel is lower than the lower threshold, then it is rejected
if a pixel is in between, then it is accepted only if it is attached to a pixel higher than the upper threshold

```python
can = cv.Canny(blur, 100, 300, 7)
fig = plt.figure(figsize=(10, 10))
fig.add_subplot(1,2,1)
plt.imshow(img)
plt.title('Original Image')
fig.add_subplot(1,2,2)
plt.imshow(can)
plt.title('Canny')
```
