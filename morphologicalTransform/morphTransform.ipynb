# Morphological opeations

## Erosion 

```python
import cv2
from matplotlib import pyplot as plt
```

```python
gray = cv2.imread("/Users/chinmayeed/Documents/TataIQ_Intern/morphologicalTransform/image.png")
image = gray
plt.imshow(gray)
```

```python
for i in range(0,3):
    eroded = cv2.erode(gray.copy(), None, iterations = i + 1)
    plt.subplot(1,3,i+1), plt.imshow(eroded), plt.title("Eroded {} times".format(i + 1))
```

## Dilation

```python
for i in range(0, 3):
    dilated = cv2.dilate(gray.copy(), None, iterations=i + 1)
    plt.subplot(1,3,i+1), plt.imshow(dilated), plt.title("Dilated {} times".format(i + 1))
```

## Opening

```python
kernelSizes = [(3, 3), (5, 5), (7, 7)]
i=1
for kernelSize in kernelSizes:
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernelSize)
    opening = cv2.morphologyEx(gray, cv2.MORPH_OPEN, kernel)
    plt.subplot(1,3,i), plt.imshow(opening), plt.title("Opening: ({}, {})".format(kernelSize[0], kernelSize[1]))
    i=i+1

```

## Closing

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernelSize)
i=1
for kernelSize in kernelSizes:
    closing = cv2.morphologyEx(gray, cv2.MORPH_CLOSE, kernel)
    plt.subplot(1,3,i), plt.imshow(closing), plt.title("Closing: ({}, {})".format(kernelSize[0], kernelSize[1]))
    i=i+1
  
```

## Morphological Gradient

```python
i=1
for kernelSize in kernelSizes:
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernelSize)
    gradient = cv2.morphologyEx(gray, cv2.MORPH_GRADIENT, kernel)
    plt.subplot(1,3,i), plt.imshow(gradient), plt.title("Gradient: ({}, {})".format(kernelSize[0], kernelSize[1]))
    i=i+1
  
```

## top hat (used to identify bright regions from a dark background)

```python
rectKernel = cv2.getStructuringElement(cv2.MORPH_RECT, (13, 5))
blackhat = cv2.morphologyEx(gray, cv2.MORPH_BLACKHAT, rectKernel)
tophat = cv2.morphologyEx(gray, cv2.MORPH_TOPHAT, rectKernel)

plt.subplot(131), plt.imshow(image), plt.title("Original")
plt.subplot(132), plt.imshow(blackhat), plt.title("Blackhat")
plt.subplot(133), plt.imshow(tophat), plt.title("Tophat")


plt.show()

```
