Image Filters using OpenCV

```python
#Importing all necessary libraries
import cv2 as cv
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

Median Blur


Removes salt and pepper noise from images

```python
#Load Image
noiseimg = cv.imread("/Users/chinmayeed/Documents/TataIQ_Intern/imageFiltering/download.jpeg")
#Remove noise
nonoiseimg = cv.medianBlur(noiseimg, 7)
#Plot images
fig = plt.figure(figsize=(10, 10))
fig.add_subplot(1,2,1)
plt.imshow(noiseimg)
plt.title('Image with noise')
fig.add_subplot(1,2,2)
plt.imshow(nonoiseimg)
plt.title('Image without noise')

```

Gaussian Filter

Gaussian filter blurs the desired area and cuts the noise with higher frequencies

```python
#Read image
gimg = cv.imread("/Users/chinmayeed/Documents/TataIQ_Intern/imageFiltering/gaus.jpeg")
#Convert to grayscale
gray = cv.cvtColor(gimg, cv.COLOR_BGR2GRAY)
#Apply Gaussian Filter
gblur = cv.GaussianBlur(gimg, (11,11), 0)
#Plot images
fig = plt.figure(figsize=(10, 10))
fig.add_subplot(1,3,1)
plt.imshow(gimg)
plt.title('Image without filter')
fig.add_subplot(1,3,2)
plt.imshow(gray)
plt.title('Image in grayscale')
fig.add_subplot(1,3,3)
plt.imshow(gblur)
plt.title('Image without noise')
```

Laplacian Filter

The Laplacian filter detects sudden intensity transitions in the image and highlights the edges.

```python
#Read the image
limg = cv.imread("/Users/chinmayeed/Documents/TataIQ_Intern/imageFiltering/lap.jpeg")
#Reducing noise
gaus =  cv.GaussianBlur(limg, (5,5), 0)
#Applying laplacian filter
dst = cv.Laplacian(gaus, cv.CV_16S, ksize=5)
#Plotting images
fig = plt.figure(figsize=(10, 10))
fig.add_subplot(1,3,1)
plt.imshow(limg)
plt.title('Original Image')
fig.add_subplot(1,3,2)
plt.imshow(gaus)
plt.title('Image without noise')
fig.add_subplot(1,3,3)
plt.imshow(dst)
plt.title('Image through Laplacian filter')
```

Sobel Operator

The Sobel operator is used to determine the approximation of the gradient of an image intensity function

```python
#Read the image
limg = cv.imread("/Users/chinmayeed/Documents/TataIQ_Intern/imageFiltering/lap.jpeg")
#Applying laplacian filter
dst = cv.Laplacian(limg, cv.CV_16S, ksize=5)
#Applying Sobel Filter
sobelx = cv.Sobel(limg,cv.CV_64F,1,0,ksize=5)
sobely = cv.Sobel(limg,cv.CV_64F,0,1,ksize=5)
#Plotting images
fig = plt.figure(figsize=(10, 10))
fig.add_subplot(1,3,1)
plt.imshow(dst)
plt.title('Laplacian')
fig.add_subplot(1,3,2)
plt.imshow(sobelx)
plt.title('Sobel X')
fig.add_subplot(1,3,3)
plt.imshow(sobely)
plt.title('Sobel Y')
```
